```python
import pandas as pd
import numpy as np
import re
```

```python
product_df = pd.read_csv('clean_skincare_product_1_1.csv')
product_df.head(3)
```

```python
review_df = pd.read_csv('Crawling Data Review Product FD.csv')
review_df.head(5)
```

### merge csv product dan review

```python
data = pd.merge(product_df, review_df, on='product_name', how='inner')
data.head(3)
```

```python
# hapus kolom yang tidak diperlukan
data.drop(['link', 'user_link', 'review_date', 'user_review', 'review_link', 'product_brand', 'date_now'],
                axis=1, inplace=True)
```

```python
# hapus data duplikat dengan mempertahankan baris pertama
data.drop_duplicates(keep='first', inplace=True)
```

### concat product_name dan variant

```python
data.product_name.duplicated().sum()
```

```python
data[['brand', 'product_name', 'user_name', 'star_rating']].duplicated().sum()
```

```python
data['product_name_variant_concat'] = data.apply(lambda x: '_'.join([x.product_name, x.variant]) 
                                                   if pd.notnull(x.variant) else x.product_name, axis=1)
```

```python
data[data.duplicated(['brand', 'product_name_variant_concat', 'user_name', 'star_rating'], keep=False)]
```

```python
data['product_name_concat'] = data.apply(lambda x: '_'.join([x.product_name, x.variant, x.subcategory]) 
                                                             if pd.notna(x.variant) 
                                                             else '_'.join([x.product_name, x.subcategory]),
                                                             axis=1)
```

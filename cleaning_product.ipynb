### concat product_name dan variant

```python
product_df.product_name.duplicated().sum()
```

pada dataset brand dan product name duplikat namun dibedakan oleh variant sehingga disini digabungkan product_name dan variant

```python
product_df['product_name_variant_concat'] = product_df.apply(lambda x: '_'.join([x.product_name, x.variant]) 
                                                   if pd.notnull(x.variant) else x.product_name, axis=1)
```

```python
product_df.head(5)
```

setelah product_name dan variant digabungkan ternyata masih terdapat data duplikat yang dimana dibedakan oleh subcategory sehingga disini digabungkan juga dengan subcategory

```python
product_df['product_name_concat'] = product_df.apply(lambda x: '_'.join([x.product_name, x.variant, x.subcategory]) 
                                                             if pd.notna(x.variant) 
                                                             else '_'.join([x.product_name, x.subcategory]),
                                                             axis=1)
```

### mapping nama product ke id numerik 

```python
# baca dataset product
data = pd.read_csv('clean_skincare_product_2_1.csv')
data
```

```python
# Impor library pandas
import pandas as pd

# Buat peta pemetaan (mapping) dari product_name ke ID numerik
unique_product_names = data['product_name_concat'].unique()
product_id_mapping = {product_name: id for id, product_name in enumerate(unique_product_names)}

# Tambahkan kolom 'product_id' ke DataFrame dengan ID numerik
data['product_id'] = data['product_name_concat'].map(product_id_mapping)
data
```

```python
# simpan dalam file csv
data.to_csv('clean_skincare_product_2_2.csv', index=False)
```
